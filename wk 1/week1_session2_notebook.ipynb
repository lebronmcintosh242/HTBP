## Part 3: String Methods - Powerful Tools for Text

Strings have built-in **methods** - special functions that work on strings. You call them using a dot (`.`) after the string.

### `.upper()` and `.lower()` - Change Case

```python
text = "Hello, Python!"

# Convert to uppercase
print(text.upper())      # HELLO, PYTHON!

# Convert to lowercase
print(text.lower())      # hello, python!

# Original text is unchanged
print(text)              # Hello, Python!
```

```python
# Useful for making comparisons case-insensitive
user_input = "YES"

# Convert to lowercase to make comparison easier
if user_input.lower() == "yes":
    print("User said yes!")
```

### `.strip()` - Remove Whitespace

```python
# Remove spaces from the beginning and end
messy = "   Hello, World!   "
clean = messy.strip()

print("Before:", messy)      # "   Hello, World!   "
print("After:", clean)        # "Hello, World!"

# Very useful for cleaning user input!
```

### `.replace()` - Replace Parts of a String

```python
# Replace one substring with another
sentence = "I love Python programming"

# Replace "Python" with "JavaScript"
new_sentence = sentence.replace("Python", "JavaScript")
print(new_sentence)  # I love JavaScript programming

# Replace spaces with underscores
with_underscores = sentence.replace(" ", "_")
print(with_underscores)  # I_love_Python_programming
```

### Other Useful String Methods

```python
text = "python is awesome"

# Capitalize first letter
print(text.capitalize())   # Python is awesome

# Title case (capitalize each word)
print(text.title())        # Python Is Awesome

# Count occurrences of a substring
print(text.count("o"))     # 2 (there are 2 'o's)

# Find position of substring
print(text.find("is"))     # 7 ("is" starts at position 7)

# Check if string starts with something
print(text.startswith("py"))  # True

# Check if string ends with something
print(text.endswith("me"))    # True
```

## Part 4: String Slicing - Extracting Parts of Strings

You can access individual characters or ranges of characters using **slicing**.

**Syntax:** `string[start:end]`
- `start` is inclusive (included)
- `end` is exclusive (not included)
- Indexing starts at 0!

```python
# String indexing
word = "Python"
#       012345  <- positions (indices)

# Access individual characters
print(word[0])
print(word[1])
print(word[5])
```

```python
# Negative indices count from the end
word = "Python"
#      -6-5-4-3-2-1  <- negative positions

print(word[-1])
print(word[-2])
print(word[-6])
```

```python
# Slicing - getting a range of characters
word = "Python"

# Get first 3 characters
print(word[0:3])

# Get characters from position 2 to 5
print(word[2:5])

# Omit start to begin from the beginning
print(word[:3])

# Omit end to go to the end
print(word[2:])

# Get everything
print(word[:])
```

```python
# Practical examples
email = "student@university.edu"

# Extract username (everything before @)
at_position = email.find("@")
username = email[:at_position]
print("Username:", username)  # student

# Extract domain (everything after @)
domain = email[at_position + 1:]
print("Domain:", domain)      # university.edu
```

```python
# Step slicing - every nth character
alphabet = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"

# Every other letter
print(alphabet[::2])   # ACEGIKMOQSUWY

# Reverse a string!
print(alphabet[::-1])  # ZYXWVUTSRQPONMLKJIHGFEDCBA
```

## Part 5: Arithmetic Operations - Math in Python

Python can perform all basic math operations. It's like a powerful calculator!

### Basic Arithmetic Operators

```python
# Addition
result = 10 + 5
print("10 + 5 =", result)

# Subtraction
result = 10 - 5
print("10 - 5 =", result)

# Multiplication
result = 10 * 5
print("10 * 5 =", result)

# Division (always returns a float)
result = 10 / 5
print("10 / 5 =", result)
print(type(result))
```

### Special Division Operators

```python
# Floor division - divides and rounds down to nearest integer
result = 10 // 3
print("10 // 3 =", result)     # 3 (not 3.333...)

# Modulus - gives the remainder after division
result = 10 % 3
print("10 % 3 =", result)      # 1 (because 10 √∑ 3 = 3 remainder 1)

# Exponentiation - raising to a power
result = 2 ** 8
print("2 ** 8 =", result)      # 256 (2 to the power of 8)
```

### Practical Uses of Modulus (%)

```python
# Check if a number is even or odd
number = 17
remainder = number % 2
print(f"{number} % 2 = {remainder}")
# If remainder is 0, number is even
# If remainder is 1, number is odd

# Check if a number is divisible by another
number = 15
if number % 5 == 0:
    print(f"{number} is divisible by 5")
```

### Operator Precedence - PEMDAS

Python follows the same order of operations you learned in math class:

1. **P**arentheses `()`
2. **E**xponents `**`
3. **M**ultiplication and **D**ivision `*`, `/`, `//`, `%` (left to right)
4. **A**ddition and **S**ubtraction `+`, `-` (left to right)

```python
# Example 1: Multiplication before addition
result = 2 + 3 * 4
print(result)  # 14, not 20
# Calculation: 3 * 4 = 12, then 2 + 12 = 14 using pemdas

# Use parentheses to change order
result = (2 + 3) * 4
print(result)  # 20
# Calculation: 2 + 3 = 5, then 5 * 4 = 20
```

```python
# Example 2: Multiple operations
result = 10 + 2 * 5 ** 2 / 4 - 3
print(result)
# Step by step:
# No parentheses so we move to exponentiation: ** ‚¨áÔ∏è
# 5 ** 2 = 25 ---> exponentiation
# 2 * 25 = 50 ---> multiplication
# 50 / 4 = 12.5 ---> division
# 10 + 12.5 = 22.5 ---> addition
# 22.5 - 3 = 19.5 ----> subtraction
```
